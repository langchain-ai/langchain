# Neptune Open Cypher QA Chain
This QA chain queries Neptune graph database using openCypher and returns human readable response

When using [Neptune Database](https://docs.aws.amazon.com/neptune/latest/userguide/intro.html) you must provide a `host` and `port` value.

```python
from langchain_community.graphs import NeptuneGraph

host = "<neptune-host>"
port = 8182
use_https = True

graph = NeptuneGraph(host=host, port=port, use_https=use_https)
```

When using [Neptune Analytics](https://docs.aws.amazon.com/neptune-analytics/latest/userguide/what-is-neptune-analytics.html) you must provide a `graphIdentifier`.

```python
from langchain_community.graphs import NeptuneAnalyticsGraph

graph_identifier = "<neptune-graph-id>"

graph = NeptuneAnalyticsGraph(graph_identifier=graph_identifier)
```

Once you have create a `NeptuneGraph` object as shown above, this can be used as an input to a `NeptuneOpenCypherQAChain`. 

```python
from langchain.chains import NeptuneOpenCypherQAChain
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(temperature=0, model="gpt-4")

chain = NeptuneOpenCypherQAChain.from_llm(llm=llm, graph=graph)

chain.run("how many outgoing routes does the Austin airport have?")
```
