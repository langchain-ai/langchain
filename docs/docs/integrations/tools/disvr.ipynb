# Disvr - Tool Discovery for AI Agents

[Disvr](https://www.disvr.top) is a tool search engine for AI agents. It helps your agent find the best external tool (MCP server, API, library) for any task.

## Setup

```bash
pip install langchain-disvr
```

Get a free API key at [disvr.top/keys](https://www.disvr.top/keys) (1,000 requests/day).

````python
from langchain_disvr import DisvrDiscoverTool, DisvrToolDetails

discover = DisvrDiscoverTool(api_key="dsvr_your_key_here")
details = DisvrToolDetails(api_key="dsvr_your_key_here")

# Find tools for a task
result = discover.invoke({"need": "translate Chinese to Thai"})
print(result)
````

## Use with an Agent

Disvr tools work seamlessly with LangChain agents. The agent will automatically use `disvr_discover` when it needs to find an external tool.

````python
from langchain_openai import ChatOpenAI
from langchain.agents import AgentExecutor, create_tool_calling_agent
from langchain_core.prompts import ChatPromptTemplate

tools = [discover, details]
llm = ChatOpenAI(model="gpt-4o")
prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful assistant. Use disvr_discover to find external tools when needed."),
    ("human", "{input}"),
    ("placeholder", "{agent_scratchpad}"),
])
agent = create_tool_calling_agent(llm, tools, prompt)
executor = AgentExecutor(agent=agent, tools=tools, verbose=True)

result = executor.invoke({"input": "Find me a tool that can do OCR on scanned PDFs"})
````
